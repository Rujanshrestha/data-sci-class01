```python
import pandas as pd
```

```python
titanic_df = pd.read_csv("train.csv")
```

# 1.Bar Graph

```python
titanic_df["Survived"].value_counts()
```

```python
titanic_df["Survived"].value_counts().plot(
    kind="bar",
    title = "Count of survived passengers" 
)
```

```python
titanic_df["Survived"].value_counts().plot(
    kind="bar",
    title = "Count of survived passengers" ,
    figsize = (5,2), 
    legend = True
)
```

```python
titanic_df["Survived"].value_counts().plot(
    kind="bar",
    title = "Count of survived passengers" ,
    figsize = (2,4), 
    legend = True,
    yticks = list(range(0,400,50)),
    xlabel = "Survived (0=No, 1=Yes)",
    ylabel = "No. of passenger",
    grid = True
)
```

```python
titanic_df["Survived"].value_counts().plot(
    kind="bar",
    title = "Count of survived passengers" ,
    figsize = (2,6), 
    legend = True,
    yticks = list(range(0,400,50)),
    xlabel = "Survived (0=No, 1=Yes)",
    ylabel = "No. of passenger",
    grid = True,
    colormap = "Dark2"
)
```

```python
group_by_gender = titanic_df[["Sex","Survived"]].groupby(by="Sex")
```

```python
group_by_gender["Survived"].value_counts()
```

```python
pd.crosstab(titanic_df.Survived, titanic_df.Sex).plot(
     kind="bar",
    title = "Count of survived passengers" ,
    figsize = (4,4), 
    legend = True,
    yticks = list(range(0,500,50)),
    xlabel = "Survived (0=No, 1=Yes)",
    ylabel = "No. of passenger",
    grid = True,
    colormap = "Dark2"
)
```

```python
group_by_Embarked = titanic_df[["Embarked","Age"]].groupby(by="Embarked")
```

```python
group_by_Embarked["Embarked"].value_counts()
```

```python
group_by_Embarked.median().plot(kind="bar")
```

# 2.Pie Chart

```python
group_by_Embarked = titanic_df[["Embarked","Age"]].groupby(by="Embarked")
group_by_Embarked.median().plot(
    kind="pie",y="Age"
)
```

```python
pd.crosstab(titanic_df.Survived, titanic_df.Sex).plot(
    kind="pie",
    subplots = True,
    title = "count of survived passengers",
    fontsize = 10,
    figsize = (6,6),
    colormap = "Dark2"
)
```
