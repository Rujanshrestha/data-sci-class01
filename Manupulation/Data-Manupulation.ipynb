# Imputation of Age column w.r.t gender provided

```python
titanic_df_2 = pd.read_csv("train2.csv")
```

```python
# Step 1: Calculate mean age for males and females
male_mean = titanic_df_2[titanic_df_2['Sex'] == 'male']['Age'].mean()
female_mean = titanic_df_2[titanic_df_2['Sex'] == 'female']['Age'].mean()

# Step 2: Fill missing Age values with the gender-specific mean
titanic_df_2.loc[(titanic_df_2['Sex'] == 'male') & (titanic_df_2['Age'].isnull()), 'Age'] = male_mean
titanic_df_2.loc[(titanic_df_2['Sex'] == 'female') & (titanic_df_2['Age'].isnull()), 'Age'] = female_mean
```

```python
print(titanic_df_2['Age'].isna().sum())
```
